## EDA

### PAIRPLOT

```python
sns.pairplot(data,hue='Sub-Category')
```

### PLOTTING CATEGORICAL DATA

```python
feature = data[["Ship Mode","Segment","State","Region","Category","Sub-Category"]]
for feature in feature:
    sns.catplot(x=feature,kind="count", data=data)
```

#### FROM ABOVE WE CAN CONCLUDE THAT, 
#### (A)IN SHIPPING MODE STANDARD CLASS IS MORE PREFERABLE.
#### (B)THERE ARE MORE ORDERS FROM CONSUMERS.
#### (C)THE BUSINESS IS MORE ACTIVE IN EAST AND WEST REGION.
#### (D)OFFICE SUPPLIES ARE MORE IN DEMAND.


### CORRELATION OF DATA

```python
corr = data.corr()
fig, ax = plt.subplots(figsize=(10,8))         
sns.heatmap(corr,ax=ax,annot=True)
```

#### FROM THE HEATMAP WE CAN CONCLUDE THAT
#### (A)SALES AND PROFITS ARE POSITIVELY CORRELATED.
####                                                                                       (B)DISCOUNTS AND PROFITS ARE NEGATIVELY CORRELATED.

### SALES & PROFIT

```python
total_sales = round(data["Sales"].sum(),2)
total_profit = round(data["Profit"].sum(),2)
print("Total Sales is ",total_sales,"USD \nTotal Profit is ",total_profit,"USD")
```

### TOP STATES BY SALES  

```python
top_states_sales = data.groupby("State").Sales.sum().nlargest(n=10)
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(12,8))
top_states_sales.plot(kind="bar",fontsize = 15,color="blue")
plt.xlabel("States",size=14)
plt.ylabel("Sales",size=14)
plt.title("Top 10 States By Sales",size=16)
plt.show()
```

#### CALIFORNIA HAS HIGHEST SALES.

### TOP STATES BY PROFITS

```python
top_states_profits = data.groupby("State").Profit.sum().nlargest(n=10)
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(12,8))
top_states_profits.plot(kind="bar",fontsize = 15,color=["blue"])
plt.xlabel("States",size=14)
plt.ylabel("Profits",size=14)
plt.title("Top 10 States By Profits",size=16)
plt.show()
```

### BOTTOM STATES BY PROFIT

```python
bottom_states_profits = data.groupby("State").Profit.sum().nsmallest(n=10)
plt.style.use('seaborn')
plt.figure(figsize=(12,8))
bottom_states_profits.plot(kind="bar",fontsize = 15,color="red")
plt.xlabel("States",size=14)
plt.ylabel("Profits",size=14)
plt.title("Bottom 10 States By Profits",size=16)
plt.show()
```

#### CALIFORNIA AND NEWYORK ARE THE MOST PROFITABLE STATES.

### DISCOUNT

### IF THERE IS NO DISCOUNT 

```python
data_n1 = data[data['Discount']==0]
sns.relplot(x='Sales',y='Profit',data=data_n1,hue="Sub-Category")
```

### SALES BY SUB-CATEGORY

```python
sub_sales=pd.pivot_table(data_n1,index='Sub-Category',values='Sales')
sub_sales.plot(kind='bar',color="blue",figsize=(12,8))
```

#### COPIERS AND MACHINES HAVE HIGHER SALES IN COMPARISON TO OTHERS.

### PROFITS BY SUB-CATEGORY

```python
sub_profits=pd.pivot_table(data_n1,index='Sub-Category',values='Profit')
sub_profits.plot(kind='bar',color="blue",figsize=(12,8))
```

#### COPIERS AND MACHINES ALSO HAVE HIGHEST PROFIT WHEN THERE IS NO DISCOUNT.

### IF THERE IS DISCOUNT

```python
data_n2 = data[data['Discount']!=0]
sns.relplot(x='Sales',y='Profit',data=data_n2,hue="Discount")
```

#### HIGHER IS THE DISCOUNT ; LESS IS THE PROFIT i.e COMPANY IS SUFFERING LOSSES.

### SALES BY SUB-CATEGORY

```python
sub_sales_d=pd.pivot_table(data_n2,index='Sub-Category',values='Sales')
sub_sales_d.plot(kind='bar',color="blue",figsize=(12,8))
```

#### AS IN THE PREVIOUS CASE,  COPIERS AND MACHINES HAVE HIGHER SALES.

### PROFITS BY SUB-CATEGORY

```python
sub_profits_d=pd.pivot_table(data_n2,index='Sub-Category',values='Profit')
sub_profits_d.plot(kind='bar',color="blue",figsize=(12,8))
```

#### COPIERS HAS HIGHEST PROFIT AS BEFORE, BUT THERE ARE LOSSES IN HANDFUL OF SUB-CATEGORY,
#### MACHINES HAS HIGHEST LOSS .

#### FROM THE ABOVE ANALYSIS, WE CAN CONCLUDE THAT ALTHOUGH MACHINES HAVE HIGHER SALES BUT DUE TO LARGE DISCOUNT THERE IS SEVERE LOSS.
#### DISCOUNT IS  A MAJOR REASON FOR LOSS.

### REGION WISE ANALYSIS

### PROFIT BY REGION

```python
plt.figure(figsize = (10,6))
sns.barplot(data = data , x = "Region",y = "Profit")
plt.show()
```

#### PROFIT IN EAST AND WEST ARE QUITE HIGH, BUT THERE IS LESS PROFIT IN CENTRAL REGION.

```python
cen_reg = list(data.groupby("Region"))[0][1]
plt.figure(figsize = (10,5))
sns.barplot(data = cen_reg, x = "Category",y = "Profit")
plt.title("Profit in Central Region",size=15)
plt.show()
```

#### FURNITURE SECTOR IS MAKING LOSS IN CENTRAL REGION.

```python
cat_furniture = list(cen_reg.groupby("Category"))[0][1]
plt.figure(figsize=(10,6))
sns.barplot(x="State",y="Profit",data=cat_furniture)
plt.title("Profit in Central Region Furniture Sector",size=15,color="red")
plt.show()
```

#### FROM ABOVE, WE CAN SAY THAT TEXAS AND ILLINOIS ARE CONTRIBUTING TO THE LOSS SIGNIFICANTLY.

### DISCOUNT BY STATE

```python
plt.figure(figsize=(10,6))
sns.barplot(x="State",y="Discount",data=cat_furniture)
plt.title("Discount in Central Region Furniture Sector",size=15,color="red")
plt.show()
```

#### FROM THE ABOVE BAR GRAPH, WE CAN CLEARY SAY THAT HEAVY DISCOUNTS ARE THE CAUSE OF LOSSES SUFFERED IN TEXAS AND ILLINOIS.

### DISCOUNT VS PROFIT

```python
plt.figure(figsize=(10,8))
sns.lineplot(x='Discount', y='Profit', data=data, label='Discount')
plt.title("Profit vs Discount", size=15)
plt.legend()
plt.show()
```
